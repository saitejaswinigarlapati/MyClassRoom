```python
#strings
text=input()

#converting the text in to upper case
print("TEXT IN UPPER CASE")
print(text.upper())
print()

#converting text in to lower case
print('TEXT IN LOWER CASE')
print(text.lower())
print()

#word count
w=0
t=text.split()
for i in t:
    w+=1
print("No.of words in text :  ",w)

#character count
c=0
for i in (t):
    for j in (i):
        c+=1
print("No.of character's in text : ",c)

#MOST COMMON WORD

#sentence capitalization
T=text.split(".")
s=''
for i in T:
    i.capitalize()
    s+=i
print("capalitized sentence : ",s)

#VOWLE COUNT
l=['a','e','i','o','u']
vc=0
for i in text.split(" "):
    for j in i:
        if j in l:
            vc+=1
print("No.of vowles in text : ",vc)

#CONSONANT COUNT
print("No.of consonants in text : ",c-vc)

#UNIQUE WORD
for i in text.split(" "):
    c=text.split().count(i)
    if c==i:
        print("Unique word in text is : ",i)
    else:
        print("No unique word in text")
        
#
```

```python
def even(n):
    if n==0:
        return n
    elif n%2==0:
        even(n-1)
        return n
even(4)
```
