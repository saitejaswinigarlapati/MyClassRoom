```python
#Write a Python function that takes a number as a parameter and checks whether the number is prime or not.
n=int(input())
def prime(n):
    if n==1 or n==0:
        print("Not a Prime")
    elif n==2:
        print("Prime")
    else:
        c=0
        for i in range(2,n):
            if n%i==0:
                c+=1
        if c>0:
            print("Not a Prime")
        else:
            print("Prime")
prime(n)
```

```python
#write a python function to find maximum  of 3  numbers
a,b,c=map(int,input().split())
def maxnum(a,b,c):
    if a>=b and a>=c:
        print(a)
    elif b>=a and b>=c:
        print(b)
    elif c>=a and c>=b:
        print(c)
maxnum(a,b,c)
```

```python
#write a function which asks user to enter price and quantity of a product . your program and display the bill amount as price  * quantity . if bill is more than 2000 discount of 20% on bill amount should be subtracted from the bill
n=int(input("enter no of products : "))
def bill(n):
    b=0
    for i in range(n):
        price=int(input(f"enter price of {i+1} product : "))
        quantity=int(input(f"enter quantity of {i+1} product : "))
        tc=price*quantity
        b+=tc
    if b>2000:
        discount=b*0.2
        print(b-discount)
    else:
        print(b)
bill(n)
```

```python
#write a program using list to insert an integer at a user defined position by shifting the element without using built-in functions,for insertion
n=int(input("enter no.of elements in the list : "))
l=[]
for i in range(n):
    e=int(input(f"Enter element{i+1} : "))
    l.append(e)
pos=int(input("Enter the position of insection : "))
ie=int(input("Enter the element tobe inserted : "))
if 0<=pos<n+1:
    l1=l[:pos]
    l1.append(ie)
    l2=l[pos:]
    
    fl=l1+l2
    print(fl)
else:
    print("invalide index")
```

```python
#write a program using list to remove an integer at a user defined position by shifting the element without using built-in functions,for insertion
n=int(input("enter no.of elements in the list : "))
l=[]
for i in range(n):
    e=int(input(f"Enter element{i+1} : "))
    l.append(e)
pos=int(input("Enter the position that should be removed : "))
if 0<=pos<n+1:
    l1=l[:pos]
    l2=l[pos+1:]
    fl=l1+l2
    print(fl)
else:
    print("invalide index")
```

```python
#write a phthon program to remove the odd elements in the list
n = int(input("Enter no.of elements in the list : "))
l = []
dl=[]
fl=[]
for i in range(n):
    e = int(input(f"Enter element{i+1} :"))
    l.append(e)

for j in l:
    if j%2!=0 :
        dl.append(j)
    else:
        fl.append(j)
print("the list is ",l)
print("removed elements list is ",dl)
print("list after removing the odd elements is ; ",fl)
```

```python
#write a program to print duplicte numbers in a list 
n=int(input("Enter no.of elements in list :  "))
l=[]
for i in range(n):
    e=int(input(f"Enter element{i+1} : "))
    l.append(e)
dl=[]
for i in l:
    if l.count(i)>1 and i not in dl:
        dl.append(i)
print(dl)
        
```
