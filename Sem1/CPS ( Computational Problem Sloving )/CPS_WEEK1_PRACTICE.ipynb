```python
#1 Population of town
T=int(input()) # No of testcases
for i in range(T):
    x,y,z=map(int,input().split())
    print("total population of town : ",x-y+z)
```

```python
#2 mimimum number of coins
T=int(input())
for i in range(T):
    x=int(input())
    if x%10==0:
        print(int(x/10))
    elif x%5==0:
        print(int((x/10)+1))
    else:
        print(-1)
```

```python
#5  Water Cooler
T=int(input())
for i in range(T):
    X,Y,M=map(int,input().split())
    if X*M<Y:
        print("YES")
    else:
        print("NO")
        
```

```python
#7 PASS OR FAIL
T=int(input())
for i in range(T):
    N,X,P=map(int,input().split())
    if (X*3)-((N-X)*1)>=P:
        print("PASS")
    else:
        print("FAIL")
```

```python
#8 playlist
T=int(input())
for i in range(T):
    N,X=map(int,input().split())
    if X*3<=N:
        print(N//(X*3))
    elif X*3>N:
        print(0)
    
```
