```python
#1
n=int(input("Enter no.of rows : "))
for i in range(n):
  for j in range(n):
    if i==n-1 or i==0 or j==0 or j==n-1:
      print("*",end=" ")
    else:
      print(" ",end=" ")
  print()
```

```python
#3
n=int(input("Enter no.of rows : "))
for i in range(n):
  for j in range(n):
    if i==n-1 or j==n-1 or i+j==n-1:
      print("*",end=" ")
    else:
      print(" ",end=" ")
  print()
```

```python
#4
n = int(input())
for i in range(1,n+1):
    for j in range(1,i+1):
        if j == i-1:
            print("0",end = " ")
        elif j == i-3:
            print("0",end = " ")
        else:
            print("1",end = " ")
    for k in range(i,n):
        print("-",end = " ")
    print()
```

```python
#5
t=int(input("Enter no.of testcases : "))
for i in range(t):
  ct=input()
  ch=map(str,input().split())
  x=0
  for i in ch:
    if i[0]==ct[0] or i[1]==ct[1]:
      x+=1
  if x>=1:
    print("yes")
  else:
    print("no")
```

```python
#6
t=int(input())
for i in range(t):
    n=int(input())
    x,y=map(int,input().split())
    ws=abs(x-1)+abs(y-1)
    bs=abs(n-x)+abs(n-y)
    if ws<=bs:
        print("white")
        
    else:
        print("black")
```
