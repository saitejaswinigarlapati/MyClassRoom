```python
#insert an element in an array without built in functions
my_list=[]
num_elements=int(input("enter the number of elements to add :"))
for i in range(num_elements):
    elements=int(input(f'enter element{i+1}: '))
    my_list.append(elements)
print("Current list:", my_list)
element_to_insert = int(input("Enter the element to insert: "))
position = int(input("Enter the position (0-based) where you want to insert it: "))
if 0 <= position <= len(my_list):
    for i in range(len(my_list) - 1, position, -1):
        
        my_list[i] = my_list[i - 1] 
    my_list[position] = element_to_insert

    print("Updated list:", my_list)
else:
    print("Invalid position for insertion")
```

```python
def is_symmetric(matrix):
    # Check if the matrix is square
    if len(matrix) != len(matrix[0]):
        return False
    
    # Check if the matrix is equal to its transpose
    return all(matrix[i][j] == matrix[j][i] for i in range(len(matrix)) for j in range(len(matrix[0])))
m=[[1,2,3],[2,3,4],[3,4,5]]
is_symmetric(m)
```
